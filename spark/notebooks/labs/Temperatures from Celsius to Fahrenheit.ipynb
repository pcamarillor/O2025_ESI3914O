```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Examples on Map Reduce") \
    .master("spark://b129606f16d5:7077") \
    .config("spark.ui.port", "4040") \
    .getOrCreate()

sc = spark.sparkContext
sc.setLogLevel("ERROR")
```

```python
temperatures_celsius = [0, 10, 20, 30, 40]

rdd = sc.parallelize(temperatures_celsius)
```

```python
rdd.collect()
```

```python
rdd_fahrenheit = rdd.map(lambda c: (c * 9/5) + 32)
rdd_fahrenheit.collect()
```
