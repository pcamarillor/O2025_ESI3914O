```python
plays = [

("UserA", "Song1"),

("UserB", "Song1"),

("UserA", "Song2"),

("UserA", "Song1"),

("UserC", "Song3"),

("UserB", "Song2"),

("UserD", "Song1"),

("UserC", "Song1"),

("UserD", "Song3"),

]

def playlistAnalizer(play_data):
    clean_records = set(play_data)

    # unique songs / user
    user_songs = {}
    for user, song in clean_records:
        if user not in user_songs:
            user_songs[user] = set()
        user_songs[user].add(song)

    user_song_counts = {user: len(songs) for user, songs in user_songs.items()}

    # unique users / song 
    song_users = {}
    for user, song in clean_records:
        if song not in song_users:
            song_users[song] = set()
        song_users[song].add(user)

    song_popularity = {song: len(users) for song, users in song_users.items()}
    max_popularity = max(song_popularity.values())
    most_popular_songs = [song for song, count in song_popularity.items() if count == max_popularity]

    results = {
        "unique_plays": list(clean_records),
        "user_song_counts": user_song_counts,
        "song_popularity": song_popularity,
        "most_popular_songs": most_popular_songs
    }

    return results

output = playlistAnalizer(plays)

print("\nUnique Plays:")
for record in output["unique_plays"]:
    print(record)

print("\nUnique Songs per User:")
for user, count in output["user_song_counts"].items():
    print(f"{user}: {count}")

print("\nPopularity:")
for song, count in output["song_popularity"].items():
    print(f"{song}: {count}")

print("\nMost Played:")
print(output["most_popular_songs"])



```
