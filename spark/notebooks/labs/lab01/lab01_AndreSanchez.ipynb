```python
#Unique songs per user
plays = [

("UserA", "Song1"),

("UserB", "Song1"),

("UserA", "Song2"),

("UserA", "Song1"),

("UserC", "Song3"),

("UserB", "Song2"),

("UserD", "Song1"),

("UserC", "Song1"),

("UserD", "Song3"),

]

unique_plays = {}
for user, song in plays:
    if user not in unique_plays:
        unique_plays[user] = set()
    unique_plays[user].add(song)
for user, songs in unique_plays.items():
    print(f"{user} : {len(songs)} songs")
```

```python
#Remove duplicate songs from a playlist
plays = [

("UserA", "Song1"),

("UserB", "Song1"),

("UserA", "Song2"),

("UserA", "Song1"), # duplicate

("UserC", "Song3"),

("UserB", "Song2"),

("UserD", "Song1"),

("UserC", "Song1"),

("UserD", "Song3"),

]
#Remove duplicate in the list of plays
unique_plays = list(set(plays))
print(unique_plays)
#Output the unique plays
for play in unique_plays:
    print(play)
    
```

```python
#Find the most popular song from a list of user plays.
plays = [

("UserA", "Song1"),

("UserB", "Song1"),

("UserA", "Song2"),

("UserA", "Song1"), # duplicate

("UserC", "Song3"),

("UserB", "Song2"),

("UserD", "Song1"),

("UserC", "Song1"),

("UserD", "Song3"),

]
#Find the most popular song
from collections import Counter
song_counts = Counter(song for user, song in plays)
most_popular_song = song_counts.most_common(1)[0]
print(f"The most popular song is {most_popular_song[0]} with {most_popular_song[1]} plays.")
```
